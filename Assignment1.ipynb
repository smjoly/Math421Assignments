```r
titanic = read.csv("titanic.csv")
library(ggplot2)
titanic$Pclass = factor(titanic$Pclass)
titanic$Survived = factor(titanic$Survived)
```

#### Graph of survived and died separated by sex, shows how more women survived

```r
ggplot(titanic) + geom_bar(mapping = aes(x = Survived, fill = Sex), position = "dodge")
```

#### Graph of class separated by survived and died, shows how the upper class fared better

```r
ggplot(titanic) + geom_bar(mapping = aes(fill = Survived, x = Pclass), position = "dodge")
```

##### Shows the age ranges where comparably more survived or died. In particular 15 - 30 has exceptionally high death rates.

```r
ggplot(titanic) + geom_density(mapping = aes(color = Survived, x = Age))
```

#### Shows how people with more expensive tickets had a higher survival rate. Correlates to Class, but not exactly

```r
ggplot(titanic) + geom_histogram(mapping = aes(x = Fare, fill = Survived), binwidth = 30)
```

##### People with one to two relations seem more likely to survive than those who have none other aboard.

```r
ggplot(titanic) + geom_bar(mapping = aes(x = SibSp+Parch, fill = Survived))
```

##### The higher the class, the more women. Corellates with the higher survival rate for the higher class above. Women in first class had a very high survival rate.

```r
ggplot(titanic) + geom_bar(mapping = aes(x = Sex, fill = Survived), position = "dodge")+  facet_wrap(~ Pclass, nrow = 2)
```

#### First class had a comparably larger amount of people with one relation on board

```r
ggplot(titanic) + geom_bar(mapping = aes(x = SibSp+Parch, fill = Pclass))
```
